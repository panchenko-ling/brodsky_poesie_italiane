```python
import csv, re
from google.colab import files
import xml.etree.ElementTree as ET
```

```python
#Initialising the XML file (after uploading to working files)
xml_filename = "brodsky_no_tags.xml"

#Initialising the CSV files
csv_locations = "Brodsky_XML_Locations.csv"
csv_persons = "Brodsky_XML_Persons_for_script.csv"
```

```python
with open(xml_filename, encoding='utf-8') as f:
    xml_text = f.read()

print(xml_text)
```

Iteration 1: Places, ru

```python
#Reading CSV values
pairs = []
with open(csv_locations, newline='', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        value = row.get("Text (ru)", "").strip()
        key = row.get("Location key", "").strip()
        if value and key and value != "NA" and key != "NA":
            pairs.append((value, key))

print(f"Loaded {len(pairs)} (value, key) pairs from CSV.")

#Function to check if a position is inside an existing <placeName> tag
def inside_placeName(text, pos):
    """Return True if index pos is inside an existing <placeName> tag."""
    open_tags = [m.start() for m in re.finditer(r"<placeName\b[^>]*>", text)]
    close_tags = [m.start() for m in re.finditer(r"</placeName>", text)]
    for o in open_tags:
        closes = [c for c in close_tags if c > o]
        if closes:
            c = closes[0]
            if o < pos < c:
                return True
    return False

#Replacing the first eligible occurrence of each value
for value, key in pairs:
    for match in re.finditer(re.escape(value), xml_text):
        start = match.start()
        if inside_placeName(xml_text, start):
            continue
        replacement = f'<placeName xml:id="{key}">{value}</placeName>'
        xml_text = xml_text[:start] + replacement + xml_text[match.end():]
        break

#Save
with open("annotated.xml", "w", encoding="utf-8") as f:
    f.write(xml_text)
```
